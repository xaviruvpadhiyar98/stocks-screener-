```python
import pandas as pd
%matplotlib inline
```

```python
ProfitLoss = pd.read_excel('task1-screener.xlsx')
ProfitLoss
```

```python
ProfitLoss.set_index('Date', inplace=True)
```

```python
dateTime = list(set(ProfitLoss.index))
dateTime.sort()
```

```python
i = 0
for x in dateTime:
    ProfitLoss.loc[x].plot(x='index',y='Low', title=x.date())
    i += 1
    if i > 1:
        break
```

```python
for x in dateTime:
    foo = ProfitLoss.loc[[x], ['index','Low']]
    print(foo.to_string(index=False))
    break
```

```python
PL = []
for x in dateTime:
    foo = ProfitLoss.loc[[x], ['Low']]
    c = foo['Low'][0] - foo['Low'][-1]
    PL.append({
        'Date':x,
        'Profit/Loss': c
    })
PL
```

```python
pd.DataFrame(PL).to_csv('Profit-Loss-Per-Day.xlsx', index=False)
```
