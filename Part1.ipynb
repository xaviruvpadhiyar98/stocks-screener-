```python
import pandas as pd
from datetime import timedelta, date, datetime
```

```python
%matplotlib inline
```

```python
df = pd.read_excel('NIFTY25JUN2010000PE.xlsx')
df
```

```python
for x in range(len(df)):
    df.loc[x,'Datetime'] = datetime.combine(df['Date'][x],df['Time'][x])
df
```

```python
DateTimeDf = df.set_index('Datetime')
DateTimeDf
```

```python
minute = [0,15,30,45]
results = []
for x in DateTimeDf.index:
    if x.minute in minute:
        first = DateTimeDf.loc[[x],['Low ', 'Date']]
        results.append(
            {
                'index': first.index[0],
                'Low': first['Low '].values[0],
                'Date': first['Date'].values[0]
            }
        )
len(results)
```

```python
ProfitLoss = pd.DataFrame(results)
ProfitLoss.set_index('Date', inplace=True)
ProfitLoss
```

```python
ProfitLoss.to_excel('task2-screener.xlsx')
```
